<font color = green>1.- Dominio, rango y regla, a que hacen referencia. Exprese en una definición. Aplique la misma en simbología matemática. </font>

$\text{El dominio hace referencia a los valores en los cuales la función está definida, el rango es el conjunto}$ 
$\text{de llegada que posee todos los valores que puede tomar una función y la regla hace referencia }$
$\text{a lo que asocia una función de un conjunto para llegar a otro.}$

$\text{Definición: Sean A y B dos conjuntos. Una función de A en B es una regla que a cada elemento de A }$
$\text{asocia un único elemento de B, el dominio son todos los elementos del conjunto A que pueden ser }$
$\text{transformados por la regla para llegar a todos los elementos del conjunto B, esta llegada se conoce como rango.}$

$\text{Simbología:}$

$\text{Dominio:}$

$ \hspace{1cm} $ $D_f = \{x \in X | \exists_y \in Y : f(x) = y \}$

$\text{Rango: }$

$ \hspace{1cm} $ $Im_f = \{y \in Y | \exists_x \in X : f(x) = y \}$

$\text{Regla:}$

$ \hspace{1cm} $ $f: X \rightarrow Y$

$\text{Por definición sabemos que}  lim _{x \to a} f(x) = L \text{si se cumple que:}$

$\hspace{3cm}$ $ \forall \epsilon \:\: \exists \delta \: \: \textrm{ tal que }$

$\hspace{1.5cm}$ $0<|x-a|<\delta    \Rightarrow |f(x)-L| < \epsilon$

$\text{Sustituyendo nos queda:}$

$\hspace{3cm}$ $ \forall \epsilon \:\: \exists \delta \: \: \textrm{ tal que }$

$\hspace{1.5cm}$ $0<|x-1|<\delta    \Rightarrow |\frac{2x + 1}{x - 3} + \frac{3}{2}| < \epsilon$

$\hspace{1.5cm}$ $0<|x-1|<\delta    \Rightarrow \frac{1}{2}|\frac{7x - 7}{x - 2}| < \epsilon$

$\hspace{1.5cm}$ $0<|x-1|<\delta    \Rightarrow 7 \cdot|\frac{x - 1}{x - 2}| < \epsilon \cdot 2$

$\hspace{1.5cm}$ $0<|x-1|<\delta    \Rightarrow |\frac{x - 1}{x - 2}| < \epsilon \cdot \frac{2}{7}$

$\hspace{1.5cm}$ $0<|x-1|<\delta    \Rightarrow |x - 1| < \epsilon \cdot \frac{2\cdot (x - 2)}{7}$

$\text{Es decir, tomando} \delta = \epsilon \cdot \frac{2\cdot (x - 2)}{7}, \text{se verifica que  } |\frac{2x + 1}{x - 3} + \frac{3}{2}|, \text{si} |x-1| < \delta, \text{luego  } lim_{x \to 1} \frac{2x + 1}{x - 3} = -\frac{3}{2}$ 

Dar la definición de límite.

$\text{Definición: Se dice que} f \text{tiene limite en el punto} a \text{si existe un número} L \in \mathbf{R} \text{tal que verifica lo siguiente:}$

$\forall\epsilon\in\mathbf{R}^+$ $\hspace{0.5cm}$ $\exists\delta\in\mathbf{R}^+ $:
           
$\left\{ 
        \begin{array}{lcc}
             0 < | x - a | < \delta \\
             x \in I\\
             \end{array}
   \right.$
   $\Rightarrow $
   $|f(x) - L| < \epsilon$
   
$\text{Dicho número se llama límite de} f \text{en} a \text{y lo escribimos: }  lim _{x \to a} f(x) = L $
